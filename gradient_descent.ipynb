```python
trainExamples = [
    (1, 1),
    (2, 3),
    (4, 3),
]
```

```python
trainExamples[0][1]
```

```python
def phi(x):
  return np.array([1,x])
```

```python
def initial_weights_vectors():
  return np.zeros(2)
```

```python

def trainLoss(w):
  denominator = len(trainExamples)
  #print(denominator)
  sum = 0 
  for x,y in trainExamples:
    #print(x)
    val = (w.dot(phi(x)) - y)**2
    sum = sum + val 
  return (sum/denominator)
```

```python
for x,y in trainExamples:
  print(x)
```

```python
w =initial_weights_vectors()
trainLoss(w)
```

```python
def gradient_loss(w):
  denominator = len(trainExamples)
  numerator = sum(2*(w.dot(phi(x))-y)*phi(x) for x,y in trainExamples)
  #print(numerator/denominator)
  return numerator/denominator
```

```python
w =initial_weights_vectors()
gradient_loss(w)
```


# Optimization algorithm

```python
def gradientDescent():
  w = initial_weights_vectors()
  n =0.1
  for t in range(500):
    trainLoss_val = trainLoss(w)
    gradient_loss_val = gradient_loss(w)
    w =w - n * gradient_loss(w)
    print(f'epoch {t}: w = {w}, F(w) = {trainLoss_val}, gradientF = {gradient_loss_val}')
```

```python
gradientDescent()
```
