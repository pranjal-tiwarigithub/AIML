```python
import numpy as np
```

```python
trainExamples = [
    # (x, y) pairs
    ((0, 2), 1),
    ((-2, 0), 1),
    ((1, -1), -1),
]
```

```python
for x,y in trainExamples:
  print(x)
```

```python
def phi_x(x):
  return np.array(x)
```

```python
def weight_initialize():
  return np.zeros(2)
```

```python
def loss(w,trainExamples):
  loss_val = 1/len(trainExamples) * sum(max(1-(w.dot(phi_x(x)*y)),0) for x,y in trainExamples)
  return loss_val 
```

```python
def gradient(w , trainExamples):
  grad_val = 1/len(trainExamples) * sum((-1) * phi_x(x) * y if (1 - w.dot(phi_x(x)) * y > 0 ) else 0 for x,y in trainExamples)
  return grad_val
```

```python
def gradient_descent(weight_initialize , loss , gradient ):
  w = weight_initialize()
  eta = 0.1
  for t in range(200):
    loss_val = loss(w,trainExamples)
    gradient_val = gradient(w , trainExamples)
    w = w - eta * gradient_val
    print(f"epoch: {t} weights:{w} loss_val: {loss_val} gradient: {gradient_val}")
```

```python
gradient_descent(weight_initialize , loss , gradient )
```
